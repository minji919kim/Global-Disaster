```python
import pandas as pd

data = pd.read_excel('/Users/minjikim/Desktop/Data Portofolio/Global Disasters/emdat_disaster.xlsx')
df = pd.DataFrame(data, columns=['Year', 'Disaster Group', 'Disaster Subgroup', 'Disaster Type', 'Country', 'Region', 'Continent', 'Location', 'Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day', 'Total Deaths', 'No Injured', "Total Damages, Adjusted ('000 US$)"])
df
```

```python
df['Start Year'] = df['Start Year'].fillna(0)
df['Start Year'] = df['Start Year'].astype(int)
df['Start Month'] = df['Start Month'].fillna(0)
df['Start Month'] = df['Start Month'].astype(int)
df['Start Day'] = df['Start Day'].fillna(0)
df['Start Day'] = df['Start Day'].astype(int)
df['End Year'] = df['End Year'].fillna(0)
df['End Year'] = df['End Year'].astype(int)
df['End Month'] = df['End Month'].fillna(0)
df['End Month'] = df['End Month'].astype(int)
df['End Day'] = df['End Day'].fillna(0)
df['End Day'] = df['End Day'].astype(int)
df["Total Damages, Adjusted"] = df["Total Damages, Adjusted ('000 US$)"] * 1000
```

```python
df['Duration'] = 'NaN'

for i in range(0, len(df)): 
    daynum = 0 
    dates = [df['Start Year'][i], df['End Year'][i], df['Start Month'][i], df['End Month'][i], df['Start Day'][i], df['End Day'][i]]
    if (all(dates) == True): 
        if (dates[5] >= dates[4] and dates[3] >= dates[2] and dates[1] >= dates[0]): 
            df['Duration'][i] = (dates[5] - dates[4]) + (dates[3] - dates[2]) * 30 + (dates[1] - dates[0]) * 365 + 1
        elif (dates[5] >= dates[4] and dates[3] < dates[2] and dates[1] > dates[0]): 
            df['Duration'][i] = (dates[5] - dates[4]) + (dates[3] - dates[2] + 12) * 30 + (dates[1] - dates[0] -1) * 365 + 1
        elif (dates[5] < dates[4] and dates[3] > dates[2] and dates[1] >= dates[0]): 
            df['Duration'][i] = (dates[5] - dates[4] + 30) + (dates[3] - dates[2] - 1) * 30 + (dates[1] - dates[0]) * 365 + 1
        elif (dates[5] < dates[4] and dates[3] <= dates[2] and dates[1] > dates[0]): 
            df['Duration'][i] = (dates[5] - dates[4] + 30) + (dates[3] - dates[2] + 11) * 30 + (dates[1] - dates[0] -1) * 365 + 1
            
df
```

```python
df.to_excel("emdat_disaster_cleaned.xlsx")
```
